```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/bremen_forecast.csv", parse_dates=["time"], index_col="time")

# PLot temperature
df["temperature_2m"].plot(figsize=(12,4), title="Temperature (2m) in Bremen", grid=True)
plt.ylabel("C")
plt.show()

# Plot wind speed
df["wind_speed_10m"].plot(figsize=(12, 4), title="Wind Speed (10m) in Bremen", grid=True)
plt.ylabel("m/s")
plt.show()

# Plot solar radiation
plt.figure(figsize=(12, 4))
df["shortwave_radiation"].plot(label="Shortwave Radiation", color="orange")
plt.title("Solar Radiation Forecast in Bremen")
plt.ylabel("W/m")
plt.xlabel("Time")
plt.grid(True)
plt.legend()
plt.show()
```
